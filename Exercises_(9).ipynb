### Introduction:

This time we will create our own dataset with fictional numbers to describe a house market. As we are going to create random data don't try to reason of the numbers.

### Step 1. Import the necessary libraries

```python
#importing the libraris
import pandas as pd
import numpy as np
```

### Step 2. Create 3 differents Series, each of length 100, as follows:
1. The first a random number from 1 to 4
2. The second a random number from 1 to 3
3. The third a random number from 10,000 to 30,000

```python
# Create 3 different Series, each of length 100
series1 = np.random.randint(1, 5, size=100)  # Random numbers from 1 to 4
series2 = np.random.randint(1, 4, size=100)  # Random numbers from 1 to 3
series3 = np.random.randint(10000, 30001, size=100)  # Random numbers from 10,000 to 30,000

# Converting them into pandas Series
series1 = pd.Series(series1)
series2 = pd.Series(series2)
series3 = pd.Series(series3)

# Display the first few values of each Series to verify
series1.head(), series2.head(), series3.head()
```

### Step 3. Let's create a DataFrame by joinning the Series by column

```python
# create a dataframe
df = pd.DataFrame({'series1': series1, 'series2': series2, 'series3': series3})
df.head()
```

### Step 4. Change the name of the columns to bedrs, bathrs, price_sqr_meter

```python
#changing the column names
df.columns = ['bedrs', 'bathrs', 'price_sqr_meter']
df.head()
```

### Step 5. Create a one column DataFrame with the values of the 3 Series and assign it to 'bigcolumn'

```python
#creating a one column dataframe
bigcolumn = pd.concat([series1, series2, series3], axis=0)
bigcolumn = bigcolumn.to_frame()
print(bigcolumn)
```
